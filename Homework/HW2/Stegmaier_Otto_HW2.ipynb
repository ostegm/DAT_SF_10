```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

column_labels = ['Sepal Length','Sepal Width', 'Petal Length', 'Petal Width', 'Class', 'Class_label']

#Included an empty column called Class_label, to be filled later with Integer values for each class
#specifying nrows to avoid pulling in the last row which ends up as nan results

df = pd.read_csv('iris.csv', names = column_labels, nrows =150)
df
```

## Step 1: Implement KNN Scoring

```python
y = df['Class_label']
y = np.asarray(y)

#Turn pandas data frame into an array
x = df[['Sepal Length','Sepal Width', 'Petal Length','Petal Width']]
X = np.asarray(x)
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier

#split the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
myknn = KNeighborsClassifier(3).fit(X_train, y_train)
myknn.score(X_test, y_test)
```

_______________________
##Step 2: Implement Cross Validation using KFolds Method

```python
from sklearn.cross_validation import KFold
```

```python
# generic cross validation function pulled from Lab 4
#used for K folds
def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold( len(X), n_folds=k_fold,
                           indices=True, shuffle=True,
                           random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ],
                         y[ train_slice  ])

        k_score = model.score(X[ test_slice ],
                              y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold

```

```python
#Return a cross valdated result of 5 folds using a 3 nearest neighbors:
myKnn = KNeighborsClassifier(3).fit
score = cross_validate(X,y, myKnn ,5)
score
```

______
## Step 3: Determine Ideal Value for K

```python
#cross validating K scores of 1-151 (odd only)
#saving results to list to print result

knn_scores = []
for n in range(1, 99, 2):
    score = cross_validate(X, y, KNeighborsClassifier(n).fit, 5)
    knn_scores.append([n,score])
    
cross_val = pd.DataFrame(knn_scores, columns=["neighbors", "scores"])    

```

##Result: K values of 11 and 15 both return the highest average accuracy

```python
max_scores = cross_val[cross_val['scores'] == cross_val['scores'].max()]
max_scores
```

```python
%matplotlib inline
#Plotting the results of testing K values between 1 and 99
plt.plot(cross_val.neighbors, cross_val.scores)
plt.title('Plot of Accuracy with increasing values of K')
```
