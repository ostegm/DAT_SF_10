```python
#Step 1: Load the data from the json file into your python environment

import json

json_data = open("ga_hw_logins.json")
data = json.load(json_data)

```

Step 2: Convert the strings into datetime objects and append them to a list 
 
Hints:

to convert the string

```python
#Step 2: Convert the strings into datetime objects and append them to a list 
#reference for directives: http://www.tutorialspoint.com/python/time_strptime.htm

from datetime import datetime

date_time_list = []

for item in data:
    n = 0
    date_object = datetime.strptime(item, '%Y-%m-%d %H:%M:%S')
    date_time_list.append(date_object)
    n+=1

date_time_list[0:2]
```

Extra Credit: 
Using python, create a sqlite3 database and upload the data into the database.  Then 
query the data to find the date and hour with the most logins. 
 
Hints: 
‐ Import sqlite package to get python functions for working with sqlite databases  
o https://docs.python.org/2/library/sqlite3.html 
‐ Use the group by feature in SQL queries 

```python
import sqlite3
conn = sqlite3.connect('homework1.db')

conn.execute("create table HW1_ExCr(d datetime)")

print "Table created successfully"

for item in date_time_list:
    conn.execute("INSERT INTO HW1_ExCr (d) VALUES (?)",(item.strftime("%Y-%m-%d %H"),))
        
conn.commit()
print "Records created successfully"

conn.close()
```

```python
conn = sqlite3.connect('homework1.db')
print "Opened database successfully"

cursor = conn.execute("SELECT d, count(d) from HW1_ExCr group by d order by -count(d)")
rows = cursor.fetchall()

for row in rows:
    print row
    
print "Operation done successfully";
conn.close()
```
